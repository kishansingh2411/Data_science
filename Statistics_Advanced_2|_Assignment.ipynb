 1: What is hypothesis testing in statistics?

 -->Ans:  Hypothesis testing in statistics is a method to make decisions or inferences about a population based on sample data.

It helps us test an assumption (called a hypothesis) about a population parameter using evidence from data.

Null Hypothesis: The assumption that there is no effect, no difference, or status quo.
Example: "The average height of students = 170 cm."

Alternative Hypothesis : The claim we want to test; suggests an effect or difference.
Example: "The average height of students ≠ 170 cm."

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

-->Ans:  Null Hypothesis : The assumption that there is no effect, no difference, or status quo.The default assumption or “status quo.”
It is what we test against using data.

Example:
"The average exam score of students = 70."
"There is no difference in blood pressure between medicine A and medicine B."


Alternative Hypothesis : The claim we want to test; suggests an effect or difference. The claim you want to prove or test for.Suggests there is an effect, difference, or relationship. Opposite of the null hypothesis.
Example: "The average height of students ≠ 170 cm."

3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

-->Ans: The significance level (alpha) is the threshold probability we choose in advance to decide whether to reject the null hypothesis .

It represents the risk of making a Type I error → rejecting null hypothesis when it is actually true.

Common values: 0.05 (5%), 0.01 (1%), 0.10 (10%).

4. What are Type I and Type II errors? Give examples of each.

-->Ans:
Type I Error (False Positive) -> Rejecting the null hypothesis when it is actually true.
In simple words: We see an effect/difference when none exists.

Probability of Type I Error = significance level (alpha).
e.g.:
A medical test says a healthy person has a disease.
In court: convicting an innocent person.
A factory claims battery life is 10 hours , but our test (by chance) rejects it.

Type II Error (False Negative)-- > Failing to reject the null hypothesis  when the alternative is actually true.
In simple words: We miss an effect/difference that actually exists.
Probability of Type II Error = Beta

e.g.:

A medical test says a sick person is healthy.
In court: letting a guilty person go free.
A new medicine actually lowers blood pressure , but our test concludes there’s no difference

 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

-->Ans:  Z-test
-> A hypothesis test used when:
-> The population standard deviation (sigma) is known.
->The sample size is large (n ≥ 30, Central Limit Theorem applies).
Assumes data follows a normal distribution.
Uses the Z-distribution (standard normal distribution, mean = 0, SD = 1).

e.g. :
Testing if the average height of students = 170 cm, when population variance is known and sample size is large.


T-test
->A hypothesis test used when:
-> The population standard deviation (sigma) is unknown (we estimate it using sample SD).
->The sample size is small (n < 30).
Assumes data follows a normal distribution (especially important for small samples).
Uses the t-distribution, which has fatter tails than Z-distribution (to account for extra uncertainty from estimating sigma).

e.g.:
Testing if the average exam score of 20 students = 70, when population variance is not known.

```python
#7: Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
import numpy as np
from scipy.stats import norm
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)


# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is significant evidence that the mean ≠ 50")
else:
    print("Fail to Reject H0: No significant evidence against mean = 50")
```

```python
#8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Standard error
se = sample_std / np.sqrt(n)

# 95% confidence interval using t-distribution
confidence = 0.95
ci = stats.t.interval(confidence, df=n-1, loc=sample_mean, scale=se)

print("Sample Mean:", sample_mean)
print("95% Confidence Interval for Mean:", ci)

# Step 3: Plot histogram of data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)

plt.title("Normal Distribution with 95% CI for Mean")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(50, 5, 100)

mean = np.mean(data)
std = np.std(data, ddof=1)  # sample standard deviation
z_scores = (data - mean) / std

print("First 5 Z-scores:", z_scores[:5])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score (Standard Deviations from Mean)")
plt.ylabel("Density")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.legend()
plt.show()
```
